```python
# Dependencies
import pandas as pd
import requests
import gmaps
import json
from pprint import pprint
import ipywidgets as widgets  # for use interface
from ipywidgets import interact, interact_manual

# Google developer API key
from config import gkey
from config import key
```

```python
# geocoordinates

# get latitude and longitude
base_url = "https://maps.googleapis.com/maps/api/geocode/json" # got it from https://developers.google.com/maps/documentation/geocoding/start
param = {'key' : gkey}
    
param['address'] = 'California'  # building up the url
lat_lng = requests.get(base_url, params = param).json()
pprint(lat_lng)   
result = lat_lng["results"][0]["geometry"]

#get latitude and logitude
lat = result["location"]["lat"]
print(lat)
lng = result["location"]["lng"]
```

```python
# fetch the breweries in Arizona
place_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

target_coordinates = f"{lat},{lng}"
target_search = "brewery"
target_radius = 80000
target_type = "restaurant"
target_type = "establishment"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
     "keyword": target_search,
    "rankby" : "distance",
     "type": target_type,
     "key": gkey
}

# run a request using our params dictionary
response = requests.get(place_url, params=params).json()
result = response["results"]
pprint(result)
```

```python
dict = {}
ca_brew = pd.DataFrame(dict)
i = 0  # to read the items from the list
for index in result:
#for i in len:
    #print(index)
    try:
        ca_brew.loc[i, 'Name'] = result[i]['name']  # store the first Brewery name
        ca_brew.loc[i, 'Address'] = result[i]['vicinity']  # store the first Brewery name
        ca_brew.loc[i, 'Rating'] = result[i]['rating']  # store the first Brewery name
        ca_brew.loc[i, 'Lat'] = result[i]['geometry']["location"]["lat"]  # store the first Brewery name
        ca_brew.loc[i, 'Lng'] = result[i]['geometry']["location"]["lng"]  # store the first Brewery name
        ca_brew.loc[i, 'Price Level'] = result[i]['price_level']  # store the first Brewery name
        ca_brew.loc[i, 'User Rating'] = result[i]['user_ratings_total']  # store the first Brewery name
        
        i += 1
    except:
        i += 1
  
ca_brew   
```

```python
# heatmap
# Configure gmaps with API key
gmaps.configure(api_key=gkey)

# Store 'Lat' and 'Lng' into  locations 
locations = ca_brew[["Lat", "Lng"]].astype(float)

# Convert Poverty Rate to float and store
# HINT: be sure to handle NaN values
rating = ca_brew["Rating"].astype(float)

# Create a Humidity Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=rating, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```

```python
# symbol the data

# Using the template add the marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Name}</dd>
<dt>Rating</dt><dd>{Rating}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
ca_brew_info = [info_box_template.format(**row) for index, row in ca_brew.iterrows()]
locations = ca_brew[["Lat", "Lng"]]

symbol = gmaps.symbol_layer(locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=3,
    info_box_content=[f"{brew}" for brew in ca_brew_info])

# Add the layer to the map
fig.add_layer(symbol)

# add the marker
markers = gmaps.marker_layer(locations)
fig.add_layer(markers)

# Display figure
fig
```
